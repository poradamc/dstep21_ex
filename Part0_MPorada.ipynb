```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## The data for this exercise can be found at https://data.oecd.org/healthres/health-spending.htm
## The dataset downloaded was a reduced dataset which looked only at "Total" cost as a "% of GDP" for the 
## years 1990-2020.  This reduced data set can be achieved by selecting these parameters in the dropdown menus
## on the website and downloaded "selected data".

## Load data from local drive (jupyter notebook and *.csv file are in same folder)

fname = "HC_spending.csv"
```

```python
data = pd.read_csv(fname)
```

```python
data
```

```python
## Select rows containing data for USA only
US_data = data[data["LOCATION"] == "USA"]
```

```python
US_data
```

```python
## Select independent variable, year, by extracting the "TIME" column from the reduced dataset "USA_data"
year = US_data["TIME"]
```

```python
year
```

```python
## Select dependent variable, cost, by extracting the "Value" column from the reduced dataset "USA_data"
cost = US_data["Value"]
```

```python
cost
```

```python
## Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

##  Add a line to that plot 
lin, = ax.plot(year, cost, color="k")

## Add gridlines
ax.grid(linestyle="--")

## Add labels to the axes
xlab = ax.set_xlabel("Year", fontsize=14)
ylab = ax.set_ylabel("Percent of GDP", fontsize=14)

## Set the range for the x values
ax.set_xlim(year.min(), year.max())

## Increase tick label size
ax.tick_params(axis="both", labelsize=14)
```
